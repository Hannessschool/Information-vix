```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("Assignment 1 GDP and spikes.csv", sep=";")
df.head()


##df["YearInt"].min()
##df["YearInt"].max()

ticks = range(df["YearInt"].min(), df["YearInt"].max() + 1, 5)
```

```python
plt.figure(figsize=(12, 6))
plt.bar(df["YearInt"], df["GDPFloat ($$)"])
plt.xlabel("Year")
plt.ylabel("GDP in $") # Escaping the dollar signs
plt.title("Bar Chart of Data")
plt.xticks(ticks)

events = df[df["YearInt"].isin([1974, 1992])]

for _, row in events.iterrows():
  plt.annotate(
      row["CommentsText"],
      xy=(row["YearInt"], row["GDPFloat ($$)"]),
  xytext=(0, 10),
  textcoords="offset points",
  ha="center",

  arrowprops=dict(arrowstyle="->"),
  )

plt.show()
```

1974: Major reforms, whatever they may have been, led to a sharp upturn in GDP fortunes

1986: Queen abdicates. Prince takes over. -> basically no impact

1992: War begins with nearby kingdom, led to a sharp downturn in GDP fortunes
